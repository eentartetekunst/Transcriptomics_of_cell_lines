```python
import scanpy as sc
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import scanpy.external as sce
import sys
import scipy 
```

```python
counts = pd.read_csv('preprocessed_calu3.csv', sep=',')
```

```python
genes = list(counts.columns[1:])
```

```python
counts4 = counts[counts['Unnamed: 0'].str.contains('4')]
counts4.index = counts4['Unnamed: 0']
counts4 = counts4.rename(columns={'Unnamed: 0': "genes"})
counts4.index = counts4.genes
counts4 = counts4.drop('genes', axis = 1)
counts4 = counts4.T
```

```python
counts8 = counts[counts['Unnamed: 0'].str.contains('8')]
counts8.index = counts8['Unnamed: 0']
counts8 = counts8.rename(columns={'Unnamed: 0': "genes"})
counts8.index = counts8.genes
counts8 = counts8.drop('genes', axis = 1)
counts8 = counts8.T
counts8
```

```python
counts_8_contr = pd.merge(counts8, counts4, on=counts8.index)
counts_8_contr.index=counts_8_contr['key_0'].values
counts_8_contr = counts_8_contr.drop('key_0', axis = 1)
counts_8_contr.to_csv('calu3_8hpi.csv')
```

```python
counts12 = counts[counts['Unnamed: 0'].str.contains('S2')]
counts12.index = counts12['Unnamed: 0']
counts12 = counts12.rename(columns={'Unnamed: 0': "genes"})
counts12.index = counts12.genes
counts12 = counts12.drop('genes', axis = 1)
counts12 = counts12.T
hrs12_s2 = counts12.filter(regex='12')
hrs12_s2
```

```python
counts12 = counts[counts['Unnamed: 0'].str.contains('12')]
counts12.index = counts12['Unnamed: 0']
counts12 = counts12.rename(columns={'Unnamed: 0': "genes"})
counts12.index = counts12.genes
counts12 = counts12.drop('genes', axis = 1)
counts12 = counts12.T
hrs12_mock = counts12.filter(regex='mock')
hrs12_mock
```

```python
counts12 = hrs12_s2.merge(hrs12_mock, how='inner', on=hrs12_mock.index)
counts12.index = counts12.key_0
counts12 = counts12.drop('key_0', axis=1)
```
