```python
import scanpy as sc
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
import scanpy.external as sce
import sys
import scipy 
```

```python
counts = pd.read_csv('preprocessed_h1299.csv', sep=',')
```

```python
genes = list(counts.columns[1:])
```

```python
conditions = []
for i in range(len(counts['Unnamed: 0'])):
    conditions.append(counts['Unnamed: 0'][i].split('-')[1:3])
set(tuple(row) for row in conditions)
```

```python
counts4 = counts[counts['Unnamed: 0'].str.contains('-4')]
counts4.index = counts4['Unnamed: 0']
counts4 = counts4.rename(columns={'Unnamed: 0': "genes"})
counts4.index = counts4.genes
counts4 = counts4.drop('genes', axis = 1)
counts4 = counts4.T
```

```python
counts12 = counts[counts['Unnamed: 0'].str.contains('12h')]
counts12.index = counts12['Unnamed: 0']
counts12 = counts12.rename(columns={'Unnamed: 0': "genes"})
counts12.index = counts12.genes
counts12 = counts12.drop('genes', axis = 1)
counts12 = counts12.T

counts_12_contr = pd.merge(counts12, counts4, on=counts12.index)
counts_12_contr.index=counts_12_contr['key_0'].values
counts_12_contr = counts_12_contr.drop('key_0', axis = 1)
counts_12_contr.to_csv('h1299_12hpi.csv')

print('exp:', counts_12_contr.filter(regex='-S2').shape[1], 'mock:', counts_12_contr.filter(regex='mock').shape[1])
```
